Question:  81% accuracy. We dropped the column on cabin, converting to be ‘0 = data not found’ and ‘1 = data found’ does it improve accuracy?

 The cabin column was dropped because it had many missing values. Instead of using this column, you convert it into a binary format, (0 for missing data and 1 for available data). This change can help improve accuracy by reducing noise from missing data.


```python
def input_age(cols):
    Age = cols[0]
    Pclass = cols[1]

    if pd.isnull(Age):
        if Pclass == 1:
            return 37  # Average age for 1st class
        elif Pclass == 2:
            return 29  # Average age for 2nd class
        else:
            return 24  # Average age for 3rd class
    else:
        return Age
```

```python
train['Age'] = train[['Age', 'Pclass']].apply(input_age, axis=1)
```

```python
# Check if 'Cabin' is in the DataFrame columns
if 'Cabin' in train.columns:
    train.drop('Cabin', axis=1, inplace=True)
else:
    print("Cabin column not found in the DataFrame.")

# Continue with the rest of your code...
```

```python
train.drop(['Name', 'Ticket'], axis=1, inplace=True)
```

```python
train = pd.get_dummies(train, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
X = train.drop('Survived', axis=1)
y = train['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
# Evaluate the model performance
from sklearn.metrics import classification_report, confusion_matrix # Import necessary functions
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```
